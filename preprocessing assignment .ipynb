# check and handle outliers in atleast 3 columns in the dataset

we can check the outliers using boxplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select the columns with potential outliers
columns_to_check = ['Age', 'Fare', 'SibSp']
```

```python
# Create box plots for selected columns
for col in columns_to_check:
    sns.boxplot(x=titanic_data[col])
    plt.title(f'Box Plot for {col}')
    plt.show()
```

```python
outliers={}
```

```python
# Print the rows with potential outliers
for col, outlier_df in outliers.items():
    print(f'Potential outliers in {col}:')
    print(outlier_df)
```

using  IQR method,remove the outliers from the data

```python
for col in columns_to_check:
    Q1 = titanic_data[col].quantile(0.25)
    Q3 = titanic_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    titanic_data[col] = titanic_data[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
```

# Do min max scaling on the feature set (take'Survived' as target).

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Extract the target variable ('Survived')
y = titanic_data['Survived']

# Drop the target variable from the feature set
X = titanic_data.drop('Survived', axis=1)

# List of numerical features to be scaled
numerical_features = ['Age', 'Fare']
```

```python
scaler = MinMaxScaler()
```

```python
# Apply Min-Max scaling to the numerical features
X[numerical_features] = scaler.fit_transform(X[numerical_features])
```
