# IMPORT THE REQUIRED LIBRARIES AND IMPORT DATA

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/titanic.csv')
data
```

# PREPROCESSING STEPS

```python
data.shape
```

```python
data.info()
```

```python
data.describe
```

```python
data.isnull().sum()
```

```python
#fill 'AGE'column by imputing mean
```

```python
mean_value=data.mean()
mean_value
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data
```

```python
#Fill 'Embarked' column by imputing mode.
```

```python
data['Embarked'].fillna(data['Embarked'].mode(),inplace=True)
data
```

```python
#find unique values of column embarked
```

```python
unique_col=data['Embarked'].unique()
unique_col
```

```python
#encode the column 'EMBARKED' by one hot encoding
```

```python
dummy_data=pd.get_dummies(data['Embarked'],prefix='Embarked')
dummy_data
```

```python
data=pd.concat([data,dummy_data],axis=1)
data
```

```python
data.drop('Embarked', axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
#FeatureEngineering
#create a new column 'FAMILY SIZE' by combing 'sibsp $ parch'
```

```python
data['FamilySize']=data['SibSp']+data['Parch']
```

```python
data.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
data
```

```python
#Drop unnecessary columns
```

```python
data.columns
```

```python
data=data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
data
```

```python
#encode the column sex
```

```python
dummy_sex=pd.get_dummies(data['Sex'],prefix='Sex')
dummy_sex
```

```python
data=pd.concat([data,dummy_sex],axis=1)
data
```

```python
data.drop(['Sex'],axis=1,inplace=True)
```

```python
data
```

```python
# Scale numerical features like 'Age' and 'Fare'
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['Age', 'Fare']] = scaler.fit_transform(data[['Age', 'Fare']])
```

```python
#Split the data set into training and testing set
```

```python
X=data.drop(['Survived'],axis=1)#feature variable
y=data['Survived']#target variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42,)
```

# CREATING KNN AND SVM MODEL

# KNN MODEL

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#create and train knearest classifier
```

```python
knn_model=KNeighborsClassifier()
```

```python
knn_model.fit(X_train,y_train)
```

```python
#make predictions on the test set
```

```python
knn_predictions=knn_model.predict(X_test)
```

```python
#Evaluation using accuracy metrics
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn_accuracy=accuracy_score(y_test,knn_predictions)
```

```python
print("K-Nearest Neighbors Classifier Accuracy:", knn_accuracy)
```

# SVM MODEL

```python
from sklearn.svm import SVC
```

```python
svm_model=SVC()
```

```python
svm_model.fit(X_train,y_train)
```

```python
#make predictions on the test test
```

```python
svc_predictions=svm_model.predict(X_test)
```

```python
#Evaluation using accuracy metrics
```

```python
svc_accuracy=accuracy_score(y_test,svc_predictions)
```

```python
print("support vector classifier Accuracy:",svc_accuracy)
```

# K-FOLD AND STRATIFIED K FOLD CROSS VALIDATION

# KFOLD CROSS VALIDATION

```python
from sklearn.model_selection import KFold
```

```python
KFold_validator=KFold(10)
```

```python
for train_index,test_index in KFold_validator.split(X,y):
    print('Training Index:',train_index)
    print('Testing Index:',test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result=cross_val_score(knn_model,X,y,cv=KFold_validator)
```

```python
cv_result
```

```python
np.mean(cv_result)#average accuracy score
```
